```python
import pandas as pd
import seaborn as sns
import numpy as np 


df = pd.read_csv('tinder_data.csv')
df
```

```python
df.columns
```

```python
df = df.drop(['user_id','username'], axis=1)
```

```python
df
```

```python
df.status.unique()
```

```python
import matplotlib.pyplot as plt

value_counts = df['status'].value_counts()
colors = ['blue', 'pink', 'green']

plt.bar(value_counts.index, value_counts.values, color=colors)
ax = plt.gca()  
ax.set_facecolor('#bce2e6')
plt.grid(color='white', linestyle='-', linewidth=0.5)



plt.title('status')
plt.xlabel('Unique Values')
plt.ylabel('Counts')


plt.show()
```
